```python
import numpy as np
import os

# Define the directory where the datasets are stored
data_dir = 'outputs'

# List all .npz files in the directory
npz_files = [f for f in os.listdir(data_dir) if f.endswith('.npz')]

datasets = {}

for file in npz_files:
    file_path = os.path.join(data_dir, file)
    # Load the dataset
    data = np.load(file_path)
    datasets[file] = {key: data[key] for key in data.files}
    
    print(f"Loaded {file}:")
    for key in data.files:
        print(f"  Shape of {key}: {data[key].shape}")
    print("-" * 20)
```

```python
# Print a few example puzzles from the 9x9 dataset
if 'dataset_4x4.npz' in datasets:
    puzzles = datasets['dataset_4x4.npz']['puzzles']
    solutions = datasets['dataset_4x4.npz']['solutions']
    
    print("Example 4x4 Puzzles (First 2):")
    for i in range(2):
        print(f"Puzzle {i+1}:")
        print(puzzles[i])
        print(f"Solution {i+1}:")
        print(solutions[i])
        print("-" * 15)
else:
    print("dataset_4x4.npz not found in loaded datasets.")
```
